```
from keras.models import load_model

def load_file(file):
  with open(file, "r") as f:
    data = f.readlines()
  return data

root = "/content/drive/My Drive/TCC/Experiments"
experiments = os.listdir(root)

results = dict()
for experiment in experiments:
  results[experiment] = dict()
  dic = results[experiment]
  for exp_i in os.listdir(os.path.join(root, experiment)):
    dic[exp_i] = dict()
    exp_dict = dic[exp_i] 
    file = os.path.join(os.path.join(root, experiment, exp_i, "results.txt"))
    try:
      data = load_file(file)
    except:
      continue
    exp_dict["Train examples"] = int(data[0].split(" ")[-1][:-1])
    exp_dict["Validation examples"] = int(data[1].split(" ")[-1][:-1])
    exp_dict["Test examples"] = int(data[2].split(" ")[-1][:-1])
    exp_dict["Balanced accuracy"] = float(data[3].split(" ")[-2])
    exp_dict["F1"] = float(data[5].split(" ")[-2])
    for line in data[6:]:
      if line.startswith("Cohen Kappa"):
        exp_dict["Cohen Kappa"] = float(line.split(" ")[-2])
      elif line.startswith("AUC"):
        exp_dict["AUC"] = float(line.split(" ")[-2])
      elif line.startswith("Precision"):
        exp_dict["Precision"] = float(line.split(" ")[-2])
      elif line.startswith("Recall"):
        exp_dict["Recall"] = float(line.split(" ")[-2])

    
```
